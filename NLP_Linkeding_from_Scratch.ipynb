# regular expressions

```python
import re

re_test = 'this is a mde up string to test 2 different regex mthods'
re_test_messy = 'This          is a made up string to test 2      different regex methods'
re_test_messy1 = 'this-is-a-made/up.string*to>>>>test---2""""different-regex-methods'
```

## \W \w are for words
## \S \s are for white spaces

```python
## splitting a sentence into a list of words
re.split('\s', re_test)
```

```python
re.split('\s', re_test_messy)
```

```python
re.split('\s+', re_test_messy)
```

```python
re.split('\W+', re_test_messy1) ## searchs for  word character
```

```python
re.split('\w+', re_test_messy1) ## searchs for non word character - USE SMALL w
```

```python
re.findall('\S+', re_test_messy) ## searchs for non whitespace characters
```

```python
re.findall('\S+', re_test_messy1) ## searchs for non whitespace characters
```

```python
re.findall('\w+', re_test_messy1)
```

# replacing a specific string

```python
pep8_test = 'I try to follow PEP8 guidelines'
pep7_test = 'I try to follow PEP7 guidelines'
pep8_test2 = 'I try to follow PEEP8 guidelines'
```

```python
re.findall('[a-z]+', pep8_test)
```

```python
re.findall('[A-Z]+', pep8_test)
```

```python
re.findall('[A-Z]+[0-9]+', pep8_test)
```

```python
re.findall('[A-Z]+[0-9]+', pep8_test2)
```
